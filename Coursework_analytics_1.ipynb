```python
# Импотрируем библиотеки 
import pandas as pd
import numpy as np
import matplotlib as mp

# Импортируем библиотеку datetime для работы с датами
import datetime
from datetime import datetime, date


# Задаем некоторые опции, которые настраивают вывод
pd.set_option('display.notebook_repr_html', True)
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 200)
pd.set_option('display.width', 200)

# Открываем файлы для чтения
course = pd.read_csv("Data/courses.csv")
course_contents = pd.read_csv("Data/course_contents.csv")
progress_phases = pd.read_csv("Data/progress_phases.csv")
progresses = pd.read_csv("Data/progresses.csv")
students = pd.read_csv("Data/students.csv ")

# Переименование столбца с целью дальнейшего объединения с другим датасетом.
course.rename(columns = {'id':'course_id'}, inplace = True)
progresses.rename(columns = {'id':'progress_id'}, inplace = True)

# Объединяем датасеты в 1 общий датафрейм
progresses = pd.merge(progress_phases, progresses)
courses = pd.merge(course, course_contents)

main = pd.merge(progresses, courses)

# Удаляем не нужный столбец
del main['Unnamed: 0']

main.head()
```

## 1.6 Подсчитываем минимальный, максимальный, средний, медианный возраст студентов

```python
# Открываем файл для чтения
students = pd.read_csv("Data/students.csv ")

# Выбрал нужные колонки из датасета
students_age = students[['id', 'city', 'birthday']]

# Столбец birthday перевожу к типу datetime
students_age.loc[:, 'birthday'] = pd.to_datetime(students_age.loc[:, 'birthday'], errors = 'coerce')

# Создал переменную now с текущей датой
now = pd.Timestamp('now')

# Добовляем новую колонку с возрастом студентов
students_age['age'] = (now - students_age['birthday']).astype('<m8[Y]')

# создал 2 переменные с минимальным и максимальным занчениями
age_min = 10
age_max = 80

# Отфильтровал занчения возраста студентов
students_age = students_age[(students_age['age'] > age_min) & (students_age['age'] < age_max)]

# Подсчитываем минимальный, максимальный, средний, медианный возраст студентов
min_age = students_age['age'].min()
max_age = students_age['age'].max()
mean_age = students_age['age'].mean()
median_age = students_age['age'].median()

# Выводим полученный результат на экран
print(f'Минимальный возраст студентов :', min_age)
print(f'Максимальный возраст студентов :', max_age)
print(f'Средний возраст студентов :', mean_age)
print(f'Медианный возраст студентов :', median_age)
```

# 1.9 Горизонтальный bar-chart с оформлением

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
fig, subplot = plt.subplots()

# Записываем напраления обчения
legend_labels = stud_count2.index.to_frame()['field'].drop_duplicates().to_list()

# Создаем списки, в которых будут храниться значения осей

x_ticks2 = []
y_ticks2 = []
labels = []

# Расчитываем мединау, для дальнейшего использования на графике
med = np.median(stud_count2.students_count.values)

# Пеперебераем каждую группу и добавляем ее на график 
for group in legend_labels:
    x_ticks = stud_count2.sort_values('students_count').loc[group].index.to_frame()['title'].to_list()
    y_ticks = stud_count2.sort_values('students_count').loc[group].students_count
    subplot.barh(x_ticks, y_ticks, label = group, alpha=0.5, edgecolor='black')
    # передаем значения для нанесения подписей на график
    x_ticks2.extend(x_ticks)
    y_ticks2.extend([med] * len(y_ticks))
    labels.extend(y_ticks)
# Добавляем значения подписий на график    
for x, y in zip(x_ticks2, labels):
    label = y
    plt.annotate(label,
                (y,x),
                textcoords='offset points',
                xytext=(15, -3),
                ha="center")
# Добавляем линию медианы на график
plt.plot(y_ticks2, x_ticks2, label='Median', linestyle='--', color='red')
# Добавляем подпись и легенду на график
plt.title('Количество студентов на каждом курсе')
plt.legend(loc='best', fontsize='small', framealpha=0.1, edgecolor='w')
# Убираем рамки
plt.box(False)
plt.show()
```

## 2.1 Расчет прироста студентов на каждом курсе в каждом месяце



```python
# создаем словарь (course_id:course_title)
id_title = {course.iloc[i].course_id:course.iloc[i].title
            for i in range(df)}
```

```python
# формируем список нужных месяцев

month = [str(i) + '-' + (j < 10) * '0' + str(j) for i in range(2016, 2020) for j in range(1, 13)]
month = month[month.index('2016-03'):month.index('2019-09')]

month_growth = pd.DataFrame(columns = ('Месяц', *list(id_title.values())))
month_growth['Месяц'] = month[:-1]
```

```python
# Формируем данные прироста студентов на каждом курсе

for course in id_title.values():
    # Создаем словарь (год,месяц, прирост студентов)
    growth = dict()
    homeworks = courses[(courses.title == course) & (courses.is_homework == True)]
    n = len(homeworks)
    start_end = ((homeworks.iloc[0].module_number,homeworks.iloc[0].lesson_number),
                 (homeworks.iloc[n - 1].module_number, homeworks.iloc[n - 1].lesson_number))
    course_df = main[main.title == course]
    for i, j in enumerate(month[:-1]):
        current_month = j
        next_month = month[i + 1]
        months = lambda x: (x > current_month) & (x < next_month)
        # Отбираем кол-во студентов, которые приступили к изучению и закончили в текущем месяце
        growth[j] = len(course_df[(course_df.module_number == start_end[0][0]) & (course_df.lesson_number == start_end[0][1])
                           & (months(course_df.start_date))])
    
    month_growth[course] = growth.values()
month_growth.head(40)
```

## 2.2 График прироста студентов на каждом курсе в каждом месяце

```python
from pandas.plotting import register_matplotlib_converters as rmc

rmc()
i = 0
for course in id_title.values():
    i += 1
    fig, subplot = plt.subplots()
    fig.set_size_inches(10, 5)
    subplot.plot(month_growth['Месяц'], month_growth[course])
    subplot.set_title(f"Ежемесячный прирост студентов на курсе: '{course}'")
    subplot.set_xlabel('Месяц, Год')
    subplot.set_ylabel('Прирост')
    plt.xticks(month_growth['Месяц'], rotation = 'vertical')
```

## 2.4 Расчет количества прогрессов по выполнению домашних работ в каждом месяце за всю историю

```python
progress_in_month = pd.DataFrame(columns = ('Месяц', *list(id_title.values())))
progress_in_month['Месяц'] = month[:-1]
```

```python
# создаем функцию для формирования формата нужной даты
cut_months = lambda x: str(x)[:7]

# Добавляем новые столбцы, с упрощенной датой
main['start_months'] = main['start_date'].apply(cut_months)
main['finish_month'] = main['finish_date'].apply(cut_months)

# Отбираем валидные значения, соответсвующим выполненым домашним заданиям
current_df = main[(main.status == 'done') & (main.is_homework == True)]
for months in range(len(month) - 1):
    month_df = current_df[(month[months] >= main.start_months) & (month[months] <= main.finish_month)]
    for course in id_title.values():
        progress_in_month[course][months] = len(month_df[month_df.title == course])
progress_in_month.head(40)
```
